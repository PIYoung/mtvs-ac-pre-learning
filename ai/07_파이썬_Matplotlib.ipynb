```python
import numpy as np
```

```python
xpoints = np.array([0, 6])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 1, 8, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
ypoints = np.array([3, 1, 8, 10, 3, 7])

plt.plot(ypoints)
plt.show()
```

```python
xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 1, 8, 10])

plt.plot(xpoints, ypoints, marker='o', ms=20)
plt.show()
```

```python
plt.plot(xpoints, ypoints, marker='o', ms=20, mec='r', mfc="y") # ms: marker size, mec: marker edge color
plt.show()
```

```python
plt.plot(xpoints, ypoints, marker='o', ms=20, mec='r', mfc="#4CAF50")
plt.show()
```

```python
plt.plot(xpoints, ypoints, marker='o', ms=20, mec='blue', mfc="hotpink", linestyle='dashed')
plt.show()
```

```python
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)

font1 = {
    'family': 'serif',
    'color': 'blue',
    'size': 20
}

plt.xlabel('this is x')
plt.ylabel('this is y')
plt.title('sport', fontdict=font1, loc='left')

plt.grid(axis='x')

plt.show()
```

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x, y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 34, 1, 10])
plt.subplot(2, 1, 2)
plt.plot(x, y)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 3, 5, 7, 9])
y = x ** 2
plt.plot(x, y, 'oy-')
plt.grid(axis='y')
plt.xlabel('X-Axis', labelpad=20)
plt.title('Graph', loc='right', pad=10)
plt.show()
```
